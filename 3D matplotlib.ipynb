# Creating 3d line

```python
z = np.linspace(0,15,1000)
x = np.sin(z)
y = np.cos(z)

ax = plt.axes(projection = '3d')
ax.plot3D(x,y,z, color = 'red')
```

```python
def c(x,y):
    return np.cos(np.sqrt(x**2 + y**2))
x = np.linspace(-5,5,25)
y = np.linspace(-5,5,25)

X, Y = np.meshgrid(x,y)      # meshgrid returns the coordinate axis from coordinate vector

z = c(X,Y)
```

```python
ax = plt.axes(projection = '3d')        # first create blank 3d plot
ax.contour3D(x,y,z)
```

```python
ax = plt.axes(projection = '3d') 
ax.contour3D(x,y,z,100)             # Increasing strides 
```

```python
#To tilt the plot

ax = plt.axes(projection = '3d') 
ax.contour3D(x,y,z,100)
ax.view_init(60,35)
```

# How to create surface plot

```python
ax = plt.axes(projection = '3d')
ax.plot_surface(X,Y,z)
```

```python
ax = plt.axes(projection = '3d')
ax.plot_surface(X,Y,z, cmap = 'spring')
```

```python
# If u have to give different colour then just pass the wrong colour and in the errors u will get the various color name.
ax = plt.axes(projection = '3d')
ax.plot_surface(X,Y,z, cmap = 'winter_r')
```

```python
ax = plt.axes(projection = '3d')
ax.plot_surface(X,Y,z, cmap = 'winter_r')
```
